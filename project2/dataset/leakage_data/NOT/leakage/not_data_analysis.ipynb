```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Increase the size of the plots
plt.rcParams['figure.figsize'] = [8, 6]
plt.rcParams['figure.dpi'] = 100
```

```python
df = pd.read_csv(f'NOT_leakage_16nm_HP.pm.csv', header=None)
df
```

```python
# Rename columns
df.columns = [ "v1", "temp" ,"vdd" ,"cqload" ,"lmin" ,"Wmin" ,"toxe_n" ,"toxe_p" ,"toxm_n" ,"toxm_p" ,"toxref_n" ,"toxref_p" ,"toxp_n" ,"toxp_p" ,"xj_n" ,"xj_p" ,"ndep_n" ,"ndep_p" ,"leakage"]
df
```

```python
import seaborn as sns

#Plot correlation matrix

plt.figure(figsize=(20,20))
cor = df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Blues)
plt.show()

# Plot Correlation of leakage with other parameters
cor_target = abs(cor["leakage"])
# Remove the target column
cor_target = cor_target.drop("leakage")
# Take square root of the correlation values
cor_target = cor_target.apply(np.sqrt).sort_values(ascending=False)
#Plot correlation with target
plt.figure(figsize=(10,10))
cor_target.plot(kind='bar')
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest,  mutual_info_regression

X = df.drop(columns=['leakage'])
y = df['leakage']
# Select the best features
selector = SelectKBest(mutual_info_regression, k=5)
X_new = selector.fit_transform(X, y)

# Get the selected features
selected_features = X.columns[selector.get_support()]
selected_features
```
