```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Increase the size of the plots
plt.rcParams['figure.figsize'] = [8, 6]
plt.rcParams['figure.dpi'] = 100
```

```python
df = pd.read_csv(f'AND_leakage_16nm_HP.pm.csv', header=None)
df
```

```python
# Rename columns
df.columns = [ "v1", "v2", "temp" ,"vdd" ,"cqload" ,"lmin" ,"Wmin" ,"toxe_n" ,"toxe_p" ,"toxm_n" ,"toxm_p" ,"toxref_n" ,"toxref_p" ,"toxp_n" ,"toxp_p" ,"xj_n" ,"xj_p" ,"ndep_n" ,"ndep_p" ,"leakage"]
df
```

```python
import seaborn as sns

#Plot correlation matrix

plt.figure(figsize=(20,20))
cor = df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Blues)
plt.show()

# Plot Correlation of leakage with other parameters
cor_target = abs(cor["leakage"])
# Remove the target column
cor_target = cor_target.drop("leakage")
# Take square root of the correlation values
cor_target = cor_target.apply(np.sqrt).sort_values(ascending=False)
#Plot correlation with target
plt.figure(figsize=(10,10))
cor_target.plot(kind='bar')
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest,  mutual_info_regression

X = df.drop(columns=['leakage'])
y = df['leakage']
# Select the best features
selector = SelectKBest(mutual_info_regression, k=5)
X_new = selector.fit_transform(X, y)

# Get the selected features
selected_features = X.columns[selector.get_support()]
selected_features
```

```python
# Plot 3d plot of v1, v2 vs leakage
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['v1'], df['v2'], df['leakage'], c='r', marker='o')
ax.set_xlabel('v1')
ax.set_ylabel('v2')
ax.set_zlabel('leakage')
plt.show()

# Take average of leakage for each v1, v2 pair and plot the average leakage
\
df_avg = df.groupby(['v1', 'v2']).mean().reset_index()

v1 = df_avg['v1'].to_numpy()
v2 = df_avg['v2'].to_numpy()
leakage_avg = df_avg['leakage'].to_numpy()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.stem3D(v1, v2, leakage_avg, linefmt='r-', markerfmt='ro', basefmt='r-')
ax.set_xlabel('v1')
ax.set_ylabel('v2')
ax.set_zlabel('leakage')
plt.show()

```

```python
X = df.drop('leakage', axis=1)
y = df['leakage']

# Apply feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
X_pca = pca.fit_transform(X_scaled)

# Plot the explained variance ratio
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio vs Number of Components')
```

```python
# Apply clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
y_kmeans = kmeans.fit_predict(X_scaled)

# Plot the clusters in 3d
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y_kmeans, cmap='viridis')
ax.set_xlabel('PCA 1')
ax.set_ylabel('PCA 2')
ax.set_zlabel('PCA 3')
plt.title('Clusters in 3D')

plt.show()
```

```python
sns.scatterplot(x='lmin', y='Wmin', hue=y_kmeans, data=df)
plt.title('Clusters in 2D')
plt.show()


```

```python
# Find the most important features
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_scaled, y)
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]
# Plot the feature importances
plt.figure()
plt.bar(range(X.shape[1]), importances[indices])
plt.xticks(range(X.shape[1]), [X.columns[i] for i in indices])
plt.xticks(rotation=90)
plt.title('Feature Importances')
plt.show()
```

```python
# Create a new dataframe with the most important features
df_imp = df[[X.columns[i] for i in indices[:5]]]
df_imp['leakage'] = y
df_imp
```

```python
# Apply PCA
X = df_imp.drop('leakage', axis=1)
y = df_imp['leakage']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

# Apply clustering
kmeans = KMeans(n_clusters=4)
y_kmeans = kmeans.fit_predict(X_scaled)

# Plot the clusters in 2d
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y_kmeans)
plt.title('Clusters in 2D')
plt.show()
```
